# SVM for salary data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.utils import resample
from sklearn.svm import SVC
import warnings
warnings.filterwarnings('ignore')
```

```python
data_raw = pd.read_csv('Database/SalaryData_train(1).csv')
data_raw
```

```python
data_raw.isnull().sum()
```

```python
for i in data_raw.columns:
    if data_raw[i].dtypes == 'object':
        print(i)
        print(data_raw[i].unique())
        print('__________________________________________________________________________')
```

```python
data_greater = data_raw[data_raw['Salary'] == ' >50K']
data_less = data_raw[data_raw['Salary'] == ' <=50K']
```

```python
data_greater['Salary'] = 1
data_greater
```

```python
data_less['Salary'] = 0
data_less
```

```python
data_res_greater = resample(data_greater, n_samples = 1000, random_state = 42, replace = False)
data_res_less = resample(data_less, n_samples = 1000, replace = False)
```

```python
data_downsampled = pd.concat([data_res_greater, data_res_less])
data_downsampled
```

```python
data_test = pd.read_csv('Database/SalaryData_Test(1).csv')
data_test['Salary'] = data_test['Salary'].map({' <=50K' : 0, ' >50K' : 1})
data_test
```

```python
data_jumbo = pd.concat([data_downsampled, data_test])
data_jumbo
```

```python
data_with_dummies = pd.get_dummies(data_jumbo)
data_with_dummies
```

```python
data_test = data_with_dummies.iloc[2000:]
data_test
```

```python
data_train = data_with_dummies.iloc[0:2000,:]
data_train
```

## Optimizing SVM

```python
y = data_train['Salary']
X = data_train.drop('Salary', axis = 1)
X_scaled = scale(X)
```

```python
param_grid = {'kernel' : ['rbf'], 'gamma' : [1, 0.1, 0.001, 0.0001, 0.00001], 'C' : [0.5, 1, 100]}
optimal = GridSearchCV(SVC(), param_grid, cv = 10)
optimal.fit(X_scaled, y)
```

```python
optimal.best_params_ , optimal.best_score_ 
```

```python
vector = SVC(C = 100, gamma = 0.001)
vector.fit(X_scaled, y)
```

```python
y_test = data_test['Salary']
X_test = data_test.drop('Salary', axis = 1)
X_test_scaled = scale(X_test)
```

```python
predictions = vector.predict(X_test_scaled)
np.mean(predictions == y_test)
```

```python
confusion_matrix(y_test, predictions)
```

```python
fig , ax = plt.subplots(figsize = (15,15))
plot_confusion_matrix(vector, X_test_scaled, y_test, ax = ax)
ax.set_xlabel('Predicted Label', fontsize = 20)
ax.set_ylabel('True Label', fontsize = 20)
plt.show()
```

## Predictions applying PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit_transform(X)
```

```python
var = pca.explained_variance_ratio_
var_sum = np.cumsum(np.round(var , 4)*100)
var_sum
```

```python
scree = pd.DataFrame()
scree['Variation'] = var*100
scree['Principal Component'] = range(1 , 103)
fig = plt.subplots(figsize = (15,10))
plt.bar(x = 'Principal Component', height = 'Variation', data = scree)
plt.xlabel('Principal Components', fontsize = 20)
plt.ylabel('Variation', fontsize = 20)
plt.title('Scree Plot', fontsize = 20)
plt.grid()
```

```python
pca_values[:,:2]
```

```python
param_grid = {'kernel' : ['rbf'], 'gamma' : [1, 0.1, 0.001, 0.0001, 0.00001], 'C' : [0.5, 1, 100]}
optimal = GridSearchCV(SVC(), param_grid, cv = 10)
optimal.fit(scale(pca_values[:,:2]), y)
```

```python
optimal.best_params_ , optimal.best_score_ 
```

```python
svc = SVC(C = 0.5, gamma = 1)
svc.fit(scale(pca_values[:,:2]), y)
```

```python
X_pca_test = pca.fit_transform(X_test)
predictions = svc.predict(scale(X_pca_test[:,:2]))
np.mean(predictions == y_test)
```

```python
confusion_matrix(y_test, predictions)
```

```python
fig , ax = plt.subplots(figsize = (15,15))
plot_confusion_matrix(svc, scale(X_pca_test[:,:2]), y_test, ax = ax)
ax.set_xlabel('Predicted Label', fontsize = 20)
ax.set_ylabel('True Label', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (15,15))
plt.scatter(X_pca_test[:,:1], X_pca_test[:,1:2], c = predictions, cmap = 'rainbow')
plt.xlabel('PC1', fontsize = 20)
plt.ylabel('PC2', fontsize = 20)
plt.title('PC2 vs PC1', fontsize = 20)
plt.grid()
```
